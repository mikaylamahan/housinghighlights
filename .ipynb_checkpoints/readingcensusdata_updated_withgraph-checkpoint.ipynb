```python
# Create dat for the new column Average Net Migration"
# test_df = county_netmigration.iloc[0, 1:10]
# list to hold average net migration per county
avg_net_migration=[]

row_count=0
given_county=0
i=0
while ( given_county < county_num ):
    test_df = county_netmigration.iloc[given_county, 1:10]
    row_count=0 
    i=0
    while ( i < len(test_df)):
        row_count+=int(test_df[i].replazce(',',''))
        i += 1;
    avg_per_county = row_count/len(test_df)
    net_migration_mean = "{0:.2f}".format(avg_per_county)
    # print(net_migration_mean)
    # add this j county net_migration_mean to the list
    avg_net_migration.append(net_migration_mean)
    given_county += 1;
print(avg_net_migration)    
```

```python
# Create the summary dataframe 
sum_df = pd.DataFrame({"County":county_netmigration["County"],"Average Net Migration": avg_net_migration})
sum_df
```

```python
sum_df_v2=sum_df.astype({"Average Net Migration": "float"})
sum_df_v2
```

```python
largest_counties=sum_df_v2.nlargest(5, "Average Net Migration")
largest_counties

largest_counties=largest_counties.set_index(["County"])
largest_counties
```

```python
#sum_df.plot.bar(x="County", y="Average Net Migration", rot=0)
x_axis=len("County")
largest_counties.plot.bar(rot=0, color='rbbbb')
plt.title("Where are People Moving? Average Net Migration 2010-2018")
plt.xlabel("Counties")
plt.ylabel("Net Migration")
plt.tick_params(axis='both', labelsize=8)

#largest_counties["Colorado"].set_color('r')
# tick_locations = [value for value in x_axis]
# plt.xticks(largest_counties["County"])
```
