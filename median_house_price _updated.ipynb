```python
%matplotlib inline
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
median_sales = "median_house.csv"
```

```python
median_pd = pd.read_csv(median_sales, encoding="ISO-8859-1")
median_pd = median_pd.fillna(0)
median_pd.head()
```

# Colorado data only

```python
Median_pd = median_pd.loc[median_pd['State'] == 'CO',:]
Median_pd.head()
median_df = pd.DataFrame(Median_pd)
median_df.head()
```

```python
#sum each month to year in datafarme
y_2010 = median_df.loc[:, "2010-01":"2010-12"].mean(axis=1)
y_2011 = median_df.loc[:, "2011-01":"2011-12"].mean(axis=1)
y_2012 = median_df.loc[:, "2012-01":"2012-12"].mean(axis=1)
y_2013 = median_df.loc[:, "2013-01":"2013-12"].mean(axis=1)
y_2014 = median_df.loc[:, "2014-01":"2014-12"].mean(axis=1)
y_2015 = median_df.loc[:, "2015-01":"2015-12"].mean(axis=1)
y_2016 = median_df.loc[:, "2016-01":"2016-12"].mean(axis=1)
y_2017 = median_df.loc[:, "2017-01":"2017-12"].mean(axis=1)
y_2018 = median_df.loc[:, "2018-01":"2018-12"].mean(axis=1)

#add new lables to correspond to years
median_df['2010'] = y_2010
median_df['2011'] = y_2011
median_df['2012'] = y_2012
median_df['2013'] = y_2013
median_df['2014'] = y_2014
median_df['2015'] = y_2015
median_df['2016'] = y_2016
median_df['2017'] = y_2017
median_df['2018'] = y_2018

#print new dataframe
median_df.head()
```

```python
#Show only coulmns needed
median_house_all = median_df[['RegionName', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']]
#set index  to "Region Name"
median_house_all = pd.DataFrame(median_house_all).set_index("RegionName")
#format all columns to  two decimals
pd.options.display.float_format = "{:,.2f}".format
#print DataFrame
median_house_all.head()
```

# Top four counties with greatest people moving in

```python
median_top_4 = median_house_all.loc[['Denver County', 'El Paso County', 'Arapahoe County', "Weld County"]]
median_top_4.T.plot(marker="o")
plt.ylabel("House Cost($)")
plt.xlabel("Year")
plt.title('Median House Cost 2010-2018')
plt.plot()
```

# Top county with people moving in Vs. counties with most people moving out

```python
top_Vs_bottom = median_house_all.loc[["Eagle County", 'Denver County', 'Moffat County']]
plt.style.use('ggplot')
top_Vs_bottom.T.plot(kind='bar')
plt.plot()
```

# Top 5 least and most expensive counties 

```python
year_18 = median_df[['RegionName','2010', '2018']]
pd.options.display.float_format = "{:,.2f}".format
sort_by_18 = year_18.sort_values('2018')
#print top 5 least expensive counties 
print(sort_by_18.head(n=5))

print()

#print top 5 most expensive counties 
print(sort_by_18.tail(n=5))
```
