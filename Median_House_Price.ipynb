```python
%matplotlib inline
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
median_sales = "median_house.csv"
```

```python
median_pd = pd.read_csv(median_sales, encoding="ISO-8859-1")
median_pd.head()
```

```python
Median_pd = median_pd.loc[median_pd['State'] == 'CO',:]
Median_pd.head()
```

```python
Median_county_pd = median_pd.loc[median_pd['RegionName'] == 'Denver County',:]
Median_county_pd.head()
```

```python
region_10 = Median_county_pd[['2010-01','2010-02','2010-03', '2010-04', '2010-05', '2010-06', '2010-07','2010-08', '2010-09', '2010-10', "2010-11", "2010-12"]]

co_median_10 = region_10.mean(axis=1)
co_median_10

df1 = co_median_10.to_frame(name='House Value')
df1
df1['Year'] = 2010
df1['County'] = 'Denver'
df1
df4 = df1.astype({"House Value": "int"})
df4
```

```python
region_18 = Median_county_pd[['2018-01','2018-02','2018-03', '2018-04', '2018-05', '2018-06', '2018-07','2018-08', '2018-09', '2018-10', "2018-11", "2018-12"]]

co_median_18 = region_18.mean(axis=1)
co_median_18

df2 = co_median_18.to_frame(name='House Value')
df2
df2['Year'] = 2018
df2['County'] = 'Denver'
df2

df3 = df2.astype({"House Value": "int"})
df3
```

```python
median_df = pd.concat([df4, df3], ignore_index=True)
median_df
```
